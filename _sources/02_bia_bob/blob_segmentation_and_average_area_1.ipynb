# Blob‑Segmentierungs‑Notebook

Dieses Notebook lädt ein Hellfeld‑Bild mit synthetischen Blobs, segmentiert die Blobs mittels einfacher Schwellenwertbestimmung und verbunden‑Komponenten‑Labeling, misst die Fläche jedes detektierten Objekts und gibt die durchschnittliche Fläche aus.

## Schritte

1. Notwendige Bibliotheken importieren.
2. Das Bild `data/blobs.tif` laden.
3. Das Bild schwellen, um eine binäre Maske zu erhalten.
4. Verbundene Komponenten in der binären Maske labeln.
5. Fläche jedes Labels messen.
6. Durchschnittliche Fläche berechnen und ausgeben.

```python
# 1. Bibliotheken importieren
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.filters import threshold_otsu
from skimage.measure import label, regionprops_table
import stackview
```

```python
# 2. Bild laden
image_path = "data/blobs.tif"
image = imread(image_path)

# Rohes Bild anzeigen
stackview.insight(image)
```

```python
# 3. Bild schwellen, um eine binäre Maske zu erstellen
thresh = threshold_otsu(image)
binary = image > thresh

# Binäre Maske anzeigen
stackview.insight(binary)
```

```python
# 4. Verbundene Komponenten beschriften
labels = label(binary)

# Beschriftetes Bild (Overlay) anzeigen
stackview.insight(labels)
```

```python
# 5. Fläche jedes Objekts messen
props = regionprops_table(labels, properties=["label", "area"])
df = pd.DataFrame(props)

# Tabelle anzeigen
display(df.head())
```

```python
# 6. Durchschnittliche Fläche berechnen und ausgeben
average_area = df["area"].mean()
print(f"Durchschnittliche Blob‑Fläche: {average_area:.2f} Pixel")
```
