# Blob Segmentation Notebook

This notebook loads a bright‑field image containing synthetic blobs, segments the blobs using simple thresholding and connected component labeling, measures the area of each detected object, and prints the average area.

## Steps

1. Import required libraries.
2. Load the image `data/blobs.tif`.
3. Threshold the image to obtain a binary mask.
4. Label connected components in the binary mask.
5. Measure area of each label.
6. Compute and print the average area.

### Step 1 – Getting the tools ready (for a 10‑year‑old)

We need some helpers that can read pictures, do math, and show the results.  These helpers are Python libraries that we "import" so we can use them later.

```python
# 1. Import libraries
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.filters import threshold_otsu
from skimage.measure import label, regionprops_table
import stackview
```

### Step 2 – Loading the picture (for a 10‑year‑old)

We tell the computer where our picture file is and ask it to open the picture so we can play with it.

```python
# 2. Load the image
image_path = "data/blobs.tif"
image = imread(image_path)

# Show the raw image
stackview.insight(image)
```

### Step 3 – Making a black‑and‑white mask (for a 10‑year‑old)

We change the picture into a simple map where everything brighter than a certain value becomes white (True) and everything else becomes black (False). This helps us separate the blobs from the background.

```python
# 3. Threshold the image to create a binary mask
thresh = threshold_otsu(image)
binary = image > thresh

# Show the binary mask
stackview.insight(binary)
```

### Step 4 – Giving each blob its own number (for a 10‑year‑old)

Now we look at the black‑and‑white map and tell the computer to count each separate white region and give it a unique label (like 1, 2, 3…).

```python
# 4. Label connected components
labels = label(binary)

# Show the labeled image (overlay)
stackview.insight(labels)
```

### Step 5 – Counting how many pixels each blob has (for a 10‑year‑old)

Each blob is made of many tiny squares called pixels. We ask the computer to tell us how many pixels belong to each blob.

```python
# 5. Measure area of each object
props = regionprops_table(labels, properties=["label", "area"])
df = pd.DataFrame(props)

# Show the table
display(df.head())
```

### Step 6 – Finding the average size (for a 10‑year‑old)

We add up the areas of all blobs and divide by how many blobs there are. This gives us the "average" size, which tells us roughly how big a typical blob is.

```python
# 6. Compute and print the average area
average_area = df["area"].mean()
print(f"Average blob area: {average_area:.2f} pixels")
```
