## Import required libraries

We import the Otsu threshold function (`threshold_otsu`) and the labeling function (`label`) from **scikit‑image**. The `imread` function is used to load the image file, and **stackview** provides interactive visualization utilities.

```python
from skimage.filters import threshold_otsu
from skimage.measure import label
from skimage.io import imread
import stackview
```

## Load the image and display it

The image `blobs.tif` is read from the *data* folder and shown using `stackview.imshow`, which opens an interactive viewer.

```python
image = imread("data/blobs.tif")
stackview.imshow(image)
```

## Segment the image

1. Compute an Otsu threshold on the grayscale image.
2. Create a binary mask by thresholding the image.
3. Label the connected components in the binary mask.
4. Visualize the resulting label image with `stackview.imshow`.

This simple pipeline separates the bright blobs from the background and assigns a unique integer label to each object.

```python
threshold = threshold_otsu(image)
binary = image > threshold
labels = label(binary)

stackview.imshow(labels)
```
